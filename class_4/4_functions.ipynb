```python
food = ["bread", "milk", "eggs"]
```

```python
#Parameters and arguments
#Print (print output) vs. return (save output to function)

def groceries(x):
    print(x)
```

```python
groceries(food)
```

```python
groceries(food[1])
```

```python
# Return statement--stops the function and can be used to save output

def groceries2(x):
    return(x)
```

```python
groceries2(food)
```

```python
y = groceries2(food)
len(y)
```

```python
#What happens if we don't use return, and try to save the output if a function to a variable?

y = groceries(food)
print(y)
```

```python
z = groceries(food)
len(z)
```

```python
# But when we use return, the function saves the output so that it can be assigned to a variable

z = groceries2(food)
len(z)
```

```python
def square_list(y):
    var2 = []
    for x in y:
      var = x*x
      var2.append(var)
        #print(x*x)
    return(var2)
```

```python
# To save multiple data from a function, use a variable inside the function to store the output data

def square_list(y):
    squares = []
    for x in y:
        squares.append(x*x)
    return(squares)

square_list(numbers_2)
```

### Exercise


Given a list foods = [“eggs”, “bread”, “milk”, “cookies”]

Create a function to return all words in the list that are longer than 5 letters.

### Solution

```python
foods = ["eggs", "bread", "milk", "cookies"]

def long_strings(x):
    for food in x:
        if len(food) >= 5:
            print(food)
    return

long_strings(foods)
```
