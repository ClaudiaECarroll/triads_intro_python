Conditionals

```python
mass = 4

if mass > 3.0:
  print(mass, 'is large')
```

```python
m = 6

if m > 3.0:
    print(m, 'is large')
    print("Hello!")
else:
    print(m, 'is small')
```

```python
a=12

if a > 9.0:
    print(a, 'is HUGE')

if a > 4:
  print("sjbailj")
elif a > 3.0:
    print(a, 'is large')
else:
    print(a, 'is small')
```

```python
# Using conditionals to update a variable

velocity = 15

if velocity > 20.0:
    print('moving too fast')
else:
    print('adjusting velocity')
    velocity = 50.0
```

```python
print(velocity)
```

```python
# What will this print?

pressure = 60

if pressure > 50.0:
    pressure = 25.0
elif pressure <= 50.0:
    pressure = 0.0

print(pressure)
```

```python
groceries = ["apples", "bread", "milk"]

if "cookies" not in groceries:
  groceries.append("cookies")

print(groceries)
```

```python
# Membership operators

numbers = [2,5,8,10]

if 4 in numbers:
  print("4 is in the list")
else:
  print("4 is not in the list")
```

```python
#Alternatively:

if 4 not in numbers:
  print("4 is not in the list")
else:
  print("4 is in the list")
```

```python
#Combination operators (OR)

if 4 in numbers or 5 in numbers:
  print("4 or 5 is in the list")
```

```python
# Combination operators (AND)

if 4 in numbers and 5 in numbers:
  print("4 and 5 is in the list")
```

```python
your_department = input("What department do you work in? ").lower()

if your_department in humanities:
    print("You are a humanist")
elif your_department in sciences:
    print("You are a scientist")
elif your_department in social_sciences:
    print("You are a social scientist")
else:
    print("You are probably doing something interesting!")
```
